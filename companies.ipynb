```python
profit=pd.read_csv('1000_Companies.csv',header=None,usecols=range(4,5),skiprows=1).dropna()
profit.shape
```

```python
profit=np.array(profit)
profit
```

```python
feature = pd.read_csv('1000_companies.csv',header=None,skiprows=1,usecols=range(0,4)).dropna()
feature.shape
```

```python
data=np.array(feature)
data
```

```python
data[:,range(3,4)]
```

```python
from sklearn import preprocessing
```

```python
le =preprocessing.LabelEncoder()

for i in range (3,4):
    le.fit(data[:,i])
    data[:,i] = le.transform(data[:,i])

data[:,range(3,4)]
```

```python
data
```

```python
X=data
Y=profit
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=12)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
```

```python
reg=reg.fit(X_train,Y_train)
print("Slope:", reg.coef_[0])
```

```python
print("Intercept:",reg.intercept_)
```

```python
m=reg.coef_[0]
b=reg.intercept_
```

```python
y_pred_train=reg.predict(X_train)
y_pred_test=reg.predict(X_test)

print("Actual Train output is", Y_train)
print("Predicit Train Output is", y_pred_train)
print(" ")
print("Actual Train output is", Y_test)
print("Predicit Train Output is", y_pred_test)
```
